```python
import pandas as pd
import numpy as np

# load data
df=pd.read_csv('/content/drive/MyDrive/titanic.csv')
df.head()
```

```python
df.drop('Cabin', axis=1, inplace=True)
from sklearn.preprocessing import LabelEncoder
label_encoders = {}

# encode the data
column_names=['PassengerId',	'Pclass',	'Name',	'Sex',	'Age',	'SibSp',	'Parch',	'Ticket',	'Fare',	'Embarked']
for column in column_names:
  label_encoders [column]=LabelEncoder()
  df[column]=label_encoders[column] .fit_transform(df[column])
```

```python
from sklearn.model_selection import train_test_split
# split data in two part
data=df.drop("Survived",axis=1)
target=df["Survived"]
print(target)

# if both the shape are equal then we can fit data
print(data.shape)
print(target.shape)
```

```python
# check all null values
df.isna().sum()
```

```python
# fill all null values wid col mean values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
# df.dropna(inplace=True)
df.isna().sum()
```

```python
from sklearn.model_selection import  train_test_split
from sklearn.ensemble import RandomForestClassifier

# split the data in traning and testing model
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)

# train the model
model= RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# test the accuray of model
prediction = model.score(X_test, y_test)
print(prediction)
```

```python
# encode the testing of model
def encode_new_data(new_data_point, label_encoders, column_names):
    new_data_encoded = []
    for value, column in zip(new_data_point, column_names):
        if column in label_encoders:
            # Encode the categorical value
            value = label_encoders[column].transform([str(value)])[0]
        new_data_encoded.append(value)
    return new_data_encoded
```

```python
# testing of model
data_set_col=[892,	3,	'Kelly, Mr. James',	'male',	34.5,	0,	0,	330911,	7.8292,	'Q']
data_set=encode_new_data(data_set_col, label_encoders, column_names)
model.predict([data_set])
```
