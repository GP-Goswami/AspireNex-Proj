```python
import pandas as pd
import numpy as np


# Load your dataset
df=pd.read_csv("/content/drive/MyDrive/WA_Fn-UseC_-Telco-Customer-Churn.csv",encoding='latin-1')

print(df.head())



# Get column names as a list
column_names = df.columns.tolist()
print(column_names)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# column which have to encode
column_name = ['customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn']
label_encoders = {}

# encode the data from stirng to float
for column in column_names:
  label_encoders [column]=LabelEncoder()
  df[column]=label_encoders[column] .fit_transform(df[column])
  # label_encoders[column] = label_encoders
```

```python
# check all the null values
df.isna().sum()
```

```python
# split the data for training and testing
data=df.drop("Churn", axis=1)
target=df["Churn"]

# if both the shape are equal then we can fit data
print(data.shape)
print(target.shape)
```

```python
# split the data for train and test to fit the model
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)

# train a model:-Random Forest for testing
model = RandomForestClassifier(n_estimators=100,
                               criterion='entropy',
                               max_depth=10,
                               min_samples_split=2,
                               min_samples_leaf=5,
                               max_features='sqrt',
                               random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# accuracy of model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# encode the testing data
def encode_new_data(new_data_point, label_encoders, column_names):
    new_data_encoded = []
    for value, column in zip(new_data_point, column_names):
        if column in label_encoders:
            # Encode the categorical value
            try:
                value = label_encoders[column].transform([str(value)])[0]
            except ValueError:
                value = -1
        new_data_encoded.append(value)
    return new_data_encoded

new_data = ['7590-VHVEG','Female',0,1,0,1,'No','No phone service', 'DSL','No','No','No','No' , 'No','No', 'Month-to-month', 1, 'Electronic check',29.85,29.85]
print(len(new_data))
# Preprocess the new data point
new_data_preprocessed = encode_new_data(new_data, label_encoders, column_name)

# Predict the new data point
prediction = model.predict([new_data_preprocessed])
print("Prediction for new data point:", prediction)
```
